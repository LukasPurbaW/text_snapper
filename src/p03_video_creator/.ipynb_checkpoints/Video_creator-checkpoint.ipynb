```python
import subprocess
import os

# ========= CONFIG =========
IMAGE_FOLDER = "snapshots_result"
SNAP_SOUND = "camera_shutter.mp3"
DURATION = 0.2
OUTPUT = "final_video.mp4"
TEMP = "temp"
# ==========================

os.makedirs(TEMP, exist_ok=True)

# Get images
image_files = sorted([
    f for f in os.listdir(IMAGE_FOLDER) if f.lower().endswith(".png")
])

segment_paths = []

for i, img in enumerate(image_files):
    img_path = os.path.join(IMAGE_FOLDER, img)
    seg_path = os.path.join(TEMP, f"seg_{i}.mp4")
    segment_paths.append(seg_path)

    # Create each segment
    cmd = [
        "ffmpeg",
        "-y",
        "-loop", "1",
        "-t", str(DURATION),
        "-i", img_path,
        "-i", SNAP_SOUND,
        "-c:v", "libx264",
        "-preset", "slow",      # better compression quality
        "-crf", "15",           # lower = better quality (range 18–23 normal, 15 is very high)
        "-pix_fmt", "yuv420p",
        "-c:a", "aac",
        "-shortest",
        seg_path
    ]

    subprocess.run(cmd)

# Create concat list
concat_path = os.path.join(TEMP, "list.txt")
with open(concat_path, "w") as f:
    for seg in segment_paths:
        f.write(f"file '{os.path.abspath(seg)}'\n")

# Final concatenation
cmd_concat = [
    "ffmpeg",
    "-y",
    "-f", "concat",
    "-safe", "0",
    "-i", concat_path,
    "-c", "copy",
    OUTPUT
]

subprocess.run(cmd_concat)

print("DONE →", OUTPUT)
```
